### Desafio 3: Colocar título no histograma das idades

```
# importando a biblioteca matplotlib para fazer gráficos
import matplotlib.pyplot as plt

df['NU_IDADE'].plot( kind='hist', bins=20, figsize=(10, 7), grid=True)
plt.title('Histograma das idades dos participantes do ENEM 2019', 
          fontsize=(15), pad=15)
plt.xlabel('Idade')
plt.ylabel('Frequência');
```

### Desafio 4: Fazer um histograma de quem é treineiro e quem não é treineiro

```
# plotando dois gráficos, um abaixo do outro
fig, ax = plt.subplots(2, 1, figsize=(7, 8), dpi=100)

ax[0].hist(df.query('IN_TREINEIRO == 1')['NU_IDADE'], bins=15)
ax[0].set_title('Histograma da idade dos treineiros do ENEM 2019.',
                fontsize=(10))
ax[0].set_xlabel('Idade')
ax[0].set_ylabel('Frequência')
ax[0].grid()

ax[1].hist(df.query('IN_TREINEIRO == 0')['NU_IDADE'], bins=15)
ax[1].set_title('Histograma da idade dos não treineiros do ENEM 2019.',
                fontsize=(10))
ax[1].set_xlabel('Idade')
ax[1].set_ylabel('Frequência')
ax[1].grid()

plt.tight_layout();
```

### Desafio 5: Comparar as distribuições das provas em inglês e espanhol

```
# a titulo de curiosidade, separei os dados para verificar a quantidade de pessoas que escolheram cada idioma nesta amostra
# 0 inglês, 1 espanhol
df['TP_LINGUA'].value_counts()
```

```
# verificando valores nulos ou sem valores
print(df.query('TP_LINGUA == 0')['NU_NOTA_LC'].isnull().value_counts())
print(df.query('TP_LINGUA == 0')['NU_NOTA_LC'].isna().value_counts())
```

```
# plotando um histograma
# o dropna foi usado para evitar o erro que estava aparecendo quando plotava valores nulos
fig, ax = plt.subplots(2, 1, figsize=(10, 8))

ax[0].hist(df.query('TP_LINGUA == 0')['NU_NOTA_LC'].dropna(),
           bins=20)
ax[0].set_title('Histograma das notas da prova de linguagens e códigos dos alunos que escolheram inglês')
ax[0].set_xlabel('Notas')
ax[0].set_ylabel('Frequência')
ax[0].grid()


ax[1].hist(df.query('TP_LINGUA == 1')['NU_NOTA_LC'].dropna(),
           bins=20)
ax[1].set_title('Histograma das notas da prova de linguagens e códigos dos alunos que escolheram espanhol')
ax[1].set_xlabel('Notas')
ax[1].set_ylabel('Frequência')
ax[1].grid()

plt.tight_layout();
```

```
# plotando histograma sobreposto
df.query('TP_LINGUA == 0')['NU_NOTA_LC'].hist(bins=30, alpha=0.5, color='red')
df.query('TP_LINGUA == 1')['NU_NOTA_LC'].hist(bins=30, alpha=0.5)
plt.title('Histograma das notas da prova de linguagens e códigos')
plt.legend(['Ingês', 'Espanhol']);
```

```
# aqui utilizei o seaborn para separar a escolha da lingua estrangeira, sendo 0 inglês e 1 espanhol
import seaborn as sns

sns.set(rc={'figure.figsize':(8, 6)})
sns.set_style('darkgrid')
sns.boxplot(x='TP_LINGUA', y='NU_NOTA_LC', 
            hue='TP_LINGUA', data=df);
```

```
plt.figure(figsize=(6,5), dpi=100)
ax = sns.boxplot(x='TP_LINGUA', y='NU_NOTA_LC', 
                data=df)

ax.set_title('Boxplot das notas da prova de linguagens e códigos')
ax.set_xlabel('Idiomas')
ax.set_ylabel('Notas de Linguagem e Código')
ax.set_xticklabels(['Inglês', 'Espanhol']);
```

Comparando os boxplots, podemos perceber que, pessoas que escolheram o inglês como lingua estrangeira na prova de linguagens e códigos tiveram um desempenho melhor. O primeiro, segundo (mediana) e terceiro quartis do TP_lingua 0 (inglês), são todos superiores ao de lingua espanhola. Esta tendência também é percebida nos valores máximos e mínimo e em alguns outliers superiores.
Entretanto, no boxplot de lingua inglesa há uma maior variação da distribuição de outliers inferiores(menores que o mínimo) em relação ao de língua espanhola. 

Neste caso, acredito que o boxplot favoreceu a análise visual quando comparamos com o histograma plotado acima (tanto o histograma separado quanto o sobreposto)
