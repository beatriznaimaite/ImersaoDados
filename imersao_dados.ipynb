### Desafio 1: Plotar as medias, medianas e moda nas notas de LC e MT (matplotlib linha vertical)

```
def cria_histograma(fonte_dados, prova):
  
  media = fonte_dados[prova].mean()
  mediana = fonte_dados[prova].median()
  moda = fonte_dados[prova].mode()

  plt.figure(figsize=(12,8))

  line1 = plt.vlines(x=media, ymin=0, ymax=3000, colors="red", linestyles="dashed")
  line2 = plt.vlines(x=mediana, ymin=0, ymax=3000, color="yellow", linestyles="dashed")
  line3 = plt.vlines(x=moda, ymin=0, ymax=3000, color="black", linestyles="dashed")

  plt.xlabel("Nota da prova de" + " " + prova)
  plt.ylabel("Frequência")

  plt.figlegend((line1, line2, line3),('Média', 'Mediana', 'Moda'))
  plt.title("ENEM 2019 - Distribuição das notas da prova de" + " " + prova)

  sns.histplot(fonte_dados, x=prova, kde=True)

cria_histograma(df_sem_zero, "NU_NOTA_MT")
cria_histograma(df_sem_zero, "NU_NOTA_LC")
```

```
provas_entrada = ['NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_LC', 'NU_NOTA_REDACAO']
prova_saida = 'NU_NOTA_MT'
```

```
notas
```

```
# eliminando NAN para o modelo rodar
df_sem_zero = df_sem_zero[notas].dropna()

#dados entrada
x = df_sem_zero[provas_entrada]

#dados saida
y = df_sem_zero[prova_saida]
```

```
# separar os dados em dados de treino e dados de teste
from sklearn.model_selection import train_test_split

SEED = 4321

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.25, 
                                                    random_state=SEED)

# o train_test_split tem um random_state nele, então os números pra treino e teste sempre se alterarão.
# caso eu queira definir sempre os mesmos valores para treino, devo definir um seed!
```

```
x_train.head()
```

```
# verificando se o x de treino e o y de treino tem a mesma quantidade de linhas
print(len(x_train))
print(len(y_train))
```

```
# criando o primeiro modelo
from sklearn.svm import LinearSVR


#instanciando o modelo e definindo uma seed no random, assim os dados não são alterados todas as vezes.
modelo = LinearSVR(random_state=SEED)

# fazendo um modelo que se ajusta aos dados de treino
# obs: não é possivel rodar o modelo se no dataframe tiver valores nulos
modelo.fit(x_train, y_train)
```

```
def calcula_mse(resultados):
  media = (resultados['test_score']*-1).mean() 
  desvio_padrao = (resultados['test_score']*-1).std()
  lim_inferior = media - (2*desvio_padrao)
  lim_superior = media + (2*desvio_padrao)
  return f'Intervalo de confiança {lim_inferior} - {lim_superior}'
```
